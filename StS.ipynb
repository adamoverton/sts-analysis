```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib notebook
plt.rcParams["figure.figsize"] = (8,6)
```

```python
df = pd.read_csv('itemInteractionsByRunCharacterFloorVictory.csv', encoding='utf_16', sep=';')
df['Upgraded'] = df['ItemName'].apply(lambda x: x[-2:] == '+1')
df['Name'] =  df['ItemName'].apply(lambda x: x[:-2] if x[-2:] == '+1' else x)
#df2 = df[['Name', 'Upgraded', 'Floor']]
#df2.groupby(by=['Name', 'Upgraded'])['Floor'].count()
df.head(3)
```

```python
dfHot = pd.get_dummies(data=df, columns=['ItemInteractionName'], dtype=float)
dfHot
```

```python
dfP = pd.pivot_table(
    dfHot.loc[(df['ItemInteractionName'] == 'card_not_picked') | (df['ItemInteractionName'] == 'card_picked')],
    index=['CharacterChosen', 'Floor'],
    columns=['Upgraded', 'Name'],
    values='ItemInteractionName_card_picked',
    aggfunc=np.mean
    )

dfP
```
